```python
df['features'] = df['features'].apply(lambda x: x.split('\n'))
```

```python
df['Status'] = np.nan
df['Carpet area'] = np.nan
df['Bathrooms'] = np.nan
df['Floor'] = np.nan
df['New/Resale'] = np.nan
df['Additional Rooms'] = np.nan
df['Balconies'] = np.nan
df['Price Negotiable'] = np.nan
```

```python
df.shape
```

```python
x = []

for i in df['features']:
    item_1 = i
    
    index = item_1.index('Status')
    next_index = index + 1
    x.append(next_index)  


for j in x:
    print(df['features'][j][x[j]])
    
y=[]
for j in x:
    y.append(df['features'][j][x[j]] )
    

df['Status'] = y
```
